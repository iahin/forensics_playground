```python
from efficient_apriori import apriori
transactions = [('eggs', 'bacon', 'soup'),
                ('eggs', 'bacon', 'apple'),
                ('soup', 'bacon', 'banana')]
itemsets, rules = apriori(transactions, min_support=0.2, min_confidence=1)

# Print out every rule with 2 items on the left hand side,
# 1 item on the right hand side, sorted by lift
rules_rhs = filter(lambda rule: len(rule.lhs) == 2 and len(rule.rhs) == 1, rules)
for rule in sorted(rules_rhs, key=lambda rule: rule.lift):
  print(rule)  # Prints the rule and its confidence, support, lift, ...
```

```python
path = 'parsed_evtx.json'
loadfile = json.load(open(path))
```

```python
allkeys = set(list(itertools.chain(*loadfile)))
allkeys
```

```python
implist = [
 'Address',
'Application',
 'CallerProcessName',
'CommandLine',
 'CommandName',
 'CommandPath',
'DestAddress',
 'DestPort',
'DomainName',
'HiveName',
'HomeDirectory',
 'HomePath',
 'HostApplication',
'ImagePath',
'IpAddress',
 'IpPort',
'NewProcessName',
'NewTargetUserName',
'ParentProcessName',
'ProcessName',
'ServiceName',
'SourceAddress',
 'SourcePort',
 'SubjectDomainName',
 'SubjectUserName',
'TargetDomainName',
 'TargetServerName',
 'TargetUserName',
 'TaskName',
'User']
```

```python
df = []

for item in loadfile:
    temp = []
    for k,v in item.items():
         if k in implist:
            temp.append(v)
    if len(temp)>2:
        df.append(tuple(temp))
df
```

```python
itemsets, rules = apriori(df, min_support=0.01, min_confidence=0.1)
```

```python
# Print out every rule with 2 items on the left hand side,
# 1 item on the right hand side, sorted by lift
rules_rhs = filter(lambda rule: len(rule.lhs) == 2 and len(rule.rhs) == 1, rules)

for rule in sorted(rules_rhs, key=lambda rule: rule.lift):
  print(rule)
```
